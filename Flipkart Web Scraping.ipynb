```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.flipkart.com/search?q=mobiles+under+20000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page="+str(1)
```

```python
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
Product_Name = []
Price = []
Ratings = []
Description = []
```

```python
box = soup.find("div",class_="_1YokD2 _3Mn1Gg")
```

```python
names = box.find_all("div",class_="_4rR01T")
# print(names)
for i in names:
    name  = i.text
    Product_Name.append(name)
# print(Product_Name)
```

```python
price = box.find_all("div",class_="_30jeq3 _1_WHN1")
for i in price:
    name  = i.text
    Price.append(name)
# print(Price)
```

```python
# print(len(Price))
```

```python
rating = box.find_all("div",class_="_3LWZlK")
for i in rating:
    name = i.text
    Ratings.append(name)
# print(Ratings)
```

```python
# print(len(Ratings))
```

```python
desc = box.find_all("ul",class_="_1xgFaf")
for i in desc:
    name = i.text
    Description.append(name)
# print(Description)
```

# For different pages iteratively

```python
Product_Name = []
Price = []
Ratings = []
Description = []

for i in range(1,21):
    url = "https://www.flipkart.com/search?q=mobiles+under+20000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off&page="+str(i)
    
    box = soup.find("div",class_="_1YokD2 _3Mn1Gg")
    
    names = box.find_all("div",class_="_4rR01T")
    for i in names:
        name  = i.text
        Product_Name.append(name)
    
    price = box.find_all("div",class_="_30jeq3 _1_WHN1")
    for i in price:
        name  = i.text
        Price.append(name)

    rating = box.find_all("div",class_="_3LWZlK")
    for i in rating:
        name = i.text
        Ratings.append(name)
    
    desc = box.find_all("ul",class_="_1xgFaf")
    for i in desc:
        name = i.text
        Description.append(name)
        
        
        
df  = pd.DataFrame({"Product_Name":Product_Name, "Price":Price, "Ratings":Ratings, "Description":Description})
print(df)
```
